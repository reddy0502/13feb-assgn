When creating a custom exception in Python, it is recommended to inherit from the built-in Exception class, which is the base class for all the exceptions. This is because the Exception class provides a number of useful features and methods that make it easier to create, handle and debug exceptions.

 some reasons why we should use the Exception class while creating a custom exception:

Inheritance: Inheriting from the Exception class allows our custom exception to inherit all of its properties and methods, such as the ability to store an error message and a traceback. 

Consistency: By using the Exception class as a base class, we ensure that our custom exception is consistent with the rest of the Python language.

Debugging: When an exception is raised, Python generates a traceback that shows the sequence of function calls that led to the exception. By inheriting from Exception, our custom exception will include a traceback that makes it easier to debug errors in our code.

```python
class PrintExceptionHierarchy:
    def __init__(self, exception_class=Exception, indent=0):
        self.exception_class = exception_class
        self.indent = indent

    def print_hierarchy(self):
        print(' ' * self.indent + self.exception_class.__name__)
        for subclass in self.exception_class.__subclasses__():
            PrintExceptionHierarchy(subclass, self.indent + 2).print_hierarchy()

PrintExceptionHierarchy().print_hierarchy()
```

ImportError is a Python built-in exception that is raised when there is an error importing a module. It usually occurs when the interpreter cannot find the module you are trying to import or when there is an error while trying to load the module.

ModuleNotFoundError is a Python exception that is raised when a module is not found during the import process. This can happen if the module does not exist in the search path, which includes the current directory, the directories specified in the PYTHONPATH environment variable, and the standard library directories.
